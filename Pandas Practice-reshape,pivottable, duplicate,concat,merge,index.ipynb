3. Find the average sales for each store

```python
data = [['A','Mumbai',40000],
       ['B','Pune',45000],
       ['A','Hyderabad',50000],
       ['C','Mumbai',90000],
       ['D','Pune',89000],
       ['A','Delhi',87000],
       ['D','Hyderabad',85000],
       ['A','Pune',78000],
       ['C','Mumbai',89000],
       ['B','Pune',70000]]

store_df = pd.DataFrame(data = data, columns=['Store','Location','Sales'])
store_df
```

```python
pd.pivot_table(data=store_df,columns=(['Store']),values=['Sales'])
```

```python
store_df.groupby('Store')['Sales'].mean()
```
